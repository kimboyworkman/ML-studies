```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller #ADF 검정
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
airpassengers = pd.read_csv('./AirPassengers.csv')
data = airpassengers['#Passengers'] #Passengers 부분만 사용하도록 하겠습니다

plt.plot(data)
plt.show()
```

```python
data.head()
```

```python
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x)
    print('ADF statistics')
    print(stat)
    print('P-value')
    print(p_value)
```

```python
adf_test(data)
```

```python
data_transformed = np.log(data).diff().diff(12) #위의 순서대로 데이터를 변환하였습니다.

data_transformed = data_transformed.dropna() #차분으로 인해 생긴 데이터의 공백을 제거합니다.
```

```python
plt.plot(data_transformed)
plt.show()
```

```python
adf_test(data_transformed)
```
